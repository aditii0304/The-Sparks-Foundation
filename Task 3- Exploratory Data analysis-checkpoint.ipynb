Data Analysis

```python
#Exploratory data Analysis
plt.figure(figsize=(8,5))
sb.kdeplot(df['Sales'], color='Green', label='Sales', shade=True,bw=50)
sb.kdeplot(df['Profit'], color='Purple', label='Profit', shade=True,bw=50)
plt.xlim([-100,1000])
plt.legend()
```

Visualisation corelation matrix

```python
#Numerical data
numerical_data=df.select_dtypes(exclude=[object])
print(numerical_data.shape)
```

```python
#Finding the corelation of numerical data
corr_matrix=numerical_data.corr()
```

```python
#Displaying the corelation matrix
corr_matrix
```

```python
#Visualising the corelation matrix using heatmap
sb.heatmap(df.corr(), cmap='rocket_r',annot=True)
```

Pairplot for analyze each column
1] Based on ship mode

```python
#Based on ship mode
sb.pairplot(df,hue='Ship Mode')
```

Count plots

```python
df['Ship Mode'].value_counts()
```

```python
sb.countplot(x='Ship Mode',data=df, palette='tab10')
plt.title('Ship Mode', fontsize='30')
```

distribuition plotting sales, profit, quantity, discount

```python
fig, axs= plt.subplots(ncols=2, nrows=2, figsize=(15,15))
sb.distplot(df['Sales'], color='red', ax=axs[0][0])
sb.distplot(df['Profit'], color='green', ax=axs[0][1])
sb.distplot(df['Quantity'], color='orange', ax=axs[1][0])
sb.distplot(df['Discount'], color='blue', ax=axs[1][1])
axs[0][0].set_title('Sales Distribution', fontsize=20)
axs[0][1].set_title('Profit Distribution', fontsize=20)
axs[1][0].set_title('quantity Distribution', fontsize=20)
axs[1][1].set_title('Discount Distribution', fontsize=20)
plt.show()
```
